```python
# Creacion de una columna
df["total"]=df["individuals"]+df["family_members"]
df.head()
```

```python
df["pc_individuals"]=df["individuals"]/df["total"]
df.head()
```

## Agregaciones

```python
df.groupby('region')
```

```python
aux=df.groupby('region').sum()
aux
```

```python
df.groupby('region')[["state_pop"]].median()
```

```python
df.groupby('region')[["individuals"]].median()
```

```python
df.groupby('region')[["individuals"]].aggregate(["std","var"])
```

```python
df.groupby(['region','state']).aggregate(['min', 'median', 'max'])
```

```python
# Agregar estadísticas de agregación (mínimo, mediana, máximo) por región
df.groupby(['region','state']).aggregate(['min', 'median', 'max'])
```

```python
df.groupby('region').aggregate({"individuals":"min","family_members":"max","state_pop":"mean"})
```
