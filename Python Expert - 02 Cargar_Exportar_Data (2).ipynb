#### Transformaciones a la data

```python
tamn=df_series_ti[['Fecha','TAMN (%)']]
```

```python
tamn
```

```python
#los : hace referencia a las filas (muestrame todas las filas) y lo que esta entre [] hace referencia a las columnas
tame=df_series_ti.loc[:,['Fecha','TAMEX (%)']]
```

```python
tame
```

```python
#iloc trabaja con los indices 0 corresponde a fecha, 1 corresponde a TAMN...
tpmn=df_series_ti.iloc[:,[0,3]]
tpmn
```

```python
#iloc trabaja con los indices -1 corresponde al indice que va de derecha a izquierda, por lo tanto corresponde a TIPMEX 
tpme=df_series_ti.iloc[:,[0,-1]]
tpme
```
